```python
def prettify_feature_name(feature_name):
    """
    Converts a snake_case feature name into a prettified format with capitalized words.

    Args:
        feature_name (str): The snake_case feature name.

    Returns:
        str: The prettified feature name with capitalized words.
    """
    return ' '.join(word.capitalize() for word in feature_name.split('_'))
```
